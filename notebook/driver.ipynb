```python
class Driver:    def __init__(self, port):        print('opening device...')        self.port =  serial.Serial(port=port,                     baudrate=115200,                     bytesize=serial.EIGHTBITS,                     stopbits=serial.STOPBITS_ONE,                     timeout=0)        self.open()        print('device initialized!')    def open(self):        self.port.open()        time.sleep(3)        self.send('g91')            def close(self):        self.port.close()    def send(self, str):        tic = timeit.default_timer()        self.port.write(str + '\r\n')        self.port.flush()        return timeit.default_timer() - tic        # time.sleep(.04)    def moveY(self, val):        return self.send('g0 y' + str(val))
```
