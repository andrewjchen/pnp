```python
import cvimport cv2import numpy as npimport matplotlib.pyplot as pltimport mathimport sysimport scipyfrom scipy import statsimport scikits.statsmodels.api as smimport randomimport numpy.linalg
```

```python
image = cv2.imread('dot.ppm', cv.CV_LOAD_IMAGE_GRAYSCALE)plt.imshow(image)
```

```python
retval, thresh = cv2.threshold(image, 50, 255, cv2.THRESH_BINARY_INV)plt.imshow(thresh)
```

```python
thresh
```

```python
contours,hierarchy = cv2.findContours(thresh,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)max_area = 0for cnt in contours:    area = cv2.contourArea(cnt)    if area > max_area:        max_area = area        best_cnt = cnt
```

```python
len(contours)
```

```python
M = cv2.moments(best_cnt)cx,cy = int(M['m10']/M['m00']), int(M['m01']/M['m00'])
```

```python
result = np.zeros([480,640], dtype=float)cv2.circle(result, (cx,cy), 5, 1)
```
